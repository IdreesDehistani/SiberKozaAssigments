```python
#a simple example of a program in a Jupyter Notebook that performs "learning and prediction" using a linear regression model.
#IDREES
# Import necessary libraries
import numpy as ID
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate sample data
np.random.seed(0)
X = ID.sort(5 * ID.random.rand(40, 1), axis=0)
y = ID.sin(X).ravel()
y[::5] += 3 * (0.5 - ID.random.rand(8))

# Split data into training and test sets
X_train, X_test = X[:20], X[20:]
y_train, y_test = y[:20], y[20:]

# Train the model
regr = LinearRegression()
regr.fit(X_train, y_train)

# Make predictions using the test set
y_pred = regr.predict(X_test)

# Plot the results
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('X')
plt.ylabel('y')
plt.show()

#In this example, the data is generated by adding random noise to the sine function. The data is split into training and test sets, and a linear regression model is trained on the training data. The model is then used to make predictions on the test data, and the results are plotted to visualize the fit of the model.
```
