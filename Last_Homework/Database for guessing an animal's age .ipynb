```python
import pandas as pd
df = pd.read_csv('CrabAgePrediction.csv')
df
```

```python
#In this step we will work on Learning and predicting
# We will ask our AI to predict the Gender(sex) of Crab according to their data
from sklearn.tree import DecisionTreeClassifier

X = df.drop(columns=['Sex'])
y = df['Sex']

model = DecisionTreeClassifier()
model.fit(X.values, y)

predictions_ID = model.predict([[11.7225, 2.1450, 4.5125, 6.678715, 22.323033, 4.6784852, 7.747181, 10]])
predictions_ID
```

```python
#On this step we will Calculate the accuracy of our test
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

df = pd.read_csv('CrabAgePrediction.csv')
X = df.drop(columns=['Sex'])
y = df['Sex']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)
predictions = model.predict(X_test)

score_ID = accuracy_score(y_test, predictions)
score_ID
```

```python
#on this step we will work on persisting models
import joblib
#In the video he uses another function but it did not work for me so I used this function instead
joblib.dump(model, 'Crab_Gender.joblib')
#In this step we are trying to see if the model works the way we want it to 
model = joblib.load('Crab_Gender.joblib')
model.fit(X.values, y)
predictions = model.predict([[2.4475, 4.6750, 66.7125, 31.535715, 22.332033, 23.584852, 4.757181, 6]])
predictions
```

```python
#This is the last step where we visualize our code 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

df = pd.read_csv('CrabAgePrediction.csv')
X = df.drop(columns=['Sex'])
y = df['Sex']
    
model = DecisionTreeClassifier()
model.fit(X.values, y)

tree.export_graphviz(model, out_file='Crab_Gender.dot',
                    feature_names=['Length','Diameter','Height','Weight','Shucked Weight','Viscera Weight','Shell Weight','Age'],
                    class_names=sorted(y.unique()),
                    label='all',
                    rounded=True,
                    filled=True) 
# For further more
```
